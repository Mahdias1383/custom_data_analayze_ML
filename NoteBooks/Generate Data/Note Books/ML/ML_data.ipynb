### <div dir="rtl" style="font-family: 'Segoe UI', Tahoma, sans-serif; font-size: 20px; color: #ffffff; font-weight: bold; padding: 10px; border-radius: 8px; background: linear-gradient(135deg, #2EA149, #60a5fa); text-align: right;"> اضافه کردن ستون های درآمد و هزینه حقیقی</div>


```python
#defining inflation indicies for each year to calculate real income
inflation_index = {
    1398: 41.2,  
    1399: 47.1,
    1400: 46.2,  
    1401: 46.5   
}
df_ML['year'] = df_ML['year'].astype(int)
```

```python
# calculating real income
df_ML['real_income'] = df_ML.apply(
    lambda row: row['income'] / inflation_index.get(int(row['year']), float('nan')), axis=1
)

# calculating real cost
df_ML['real_cost'] = df_ML.apply(
    lambda row: row['cost'] / inflation_index.get(int(row['year']), float('nan')), axis=1
)

# calculating real transportaion cost
df_ML['real_trans_cost'] = df_ML.apply(
    lambda row: row['trans_cost'] / inflation_index.get(int(row['year']), float('nan')), axis=1
)
```
