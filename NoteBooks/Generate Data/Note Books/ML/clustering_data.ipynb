### <div dir="rtl" style="font-family: 'Segoe UI', Tahoma, sans-serif; font-size: 20px; color: #ffffff; font-weight: bold; padding: 10px; border-radius: 8px; background: linear-gradient(135deg, #2EA149, #60a5fa); text-align: right;"> افزودن ستون سال به مقادیر </div>

### <div dir="rtl" style="font-family: 'Segoe UI', Tahoma, sans-serif; font-size: 20px; color: #ffffff; font-weight: bold; padding: 10px; border-radius: 8px; background: linear-gradient(135deg, #2EA149, #60a5fa); text-align: right;"> ترکیب دیتافریم های یکسان </div>


```python
df_98_R = pd.concat([df[['Address', 'value']] for df in df_98_R.values()], ignore_index=True)
df_98_U = pd.concat([df[['Address', 'value']] for df in df_98_U.values()], ignore_index=True)
df_99_R = pd.concat([df[['Address', 'value']] for df in df_99_R.values()], ignore_index=True)
df_99_U = pd.concat([df[['Address', 'value']] for df in df_99_U.values()], ignore_index=True)
df_1400_R = pd.concat([df[['Address', 'value']] for df in df_1400_R.values()], ignore_index=True)
df_1400_U = pd.concat([df[['Address', 'value']] for df in df_1400_U.values()], ignore_index=True)
df_1401_R = pd.concat([df[['Address', 'value']] for df in df_1401_R.values()], ignore_index=True)
df_1401_U = pd.concat([df[['Address', 'value']] for df in df_1401_U.values()], ignore_index=True)
```

```python
df_98=pd.concat([df_98_R,df_98_U])
df_99=pd.concat([df_99_R,df_99_U])
df_1400=pd.concat([df_1400_R,df_1400_U])
df_1401=pd.concat([df_1401_R,df_1401_U])
```

### <div dir="rtl" style="font-family: 'Segoe UI', Tahoma, sans-serif; font-size: 20px; color: #ffffff; font-weight: bold; padding: 10px; border-radius: 8px; background: linear-gradient(135deg, #2EA149, #60a5fa); text-align: right;"> افزودن ستون سال به دیتافریم ها </div>


```python
df_98['year']=1398
df_99['year']=1399
df_1400['year']=1400
df_1401['year']=1401
```

### <div dir="rtl" style="font-family: 'Segoe UI', Tahoma, sans-serif; font-size: 20px; color: #ffffff; font-weight: bold; padding: 10px; border-radius: 8px; background: linear-gradient(135deg, #2EA149, #60a5fa); text-align: right;"> ترکیب دیتافریم های یکسان </div>


```python
df_costs = pd.concat([df_98,df_99,df_1400,df_1401])
df_costs
```

### <div dir="rtl" style="font-family: 'Segoe UI', Tahoma, sans-serif; font-size: 20px; color: #ffffff; font-weight: bold; padding: 10px; border-radius: 8px; background: linear-gradient(135deg, #2EA149, #60a5fa); text-align: right;"> بررسی داده ها </div>


```python
df_costs.info()
```

### <div dir="rtl" style="font-family: 'Segoe UI', Tahoma, sans-serif; font-size: 20px; color: #ffffff; font-weight: bold; padding: 10px; border-radius: 8px; background: linear-gradient(135deg, #2EA149, #60a5fa); text-align: right;"> تبدیل مقادیر غیر عددی به مقادیر عددی </div>


```python
df_costs = df_costs.apply(pd.to_numeric, errors='coerce') 
df_costs.info()
```

### <div dir="rtl" style="font-family: 'Segoe UI', Tahoma, sans-serif; font-size: 20px; color: #ffffff; font-weight: bold; padding: 10px; border-radius: 8px; background: linear-gradient(135deg, #2EA149, #60a5fa); text-align: right;"> ترکیب داده ها و جمع مقادیر </div>


```python
df_costs = df_costs.groupby(['Address','year']).agg(
    cost=pd.NamedAgg(column='value', aggfunc='sum'),
    
).reset_index()

df_costs
```

### <div dir="rtl" style="font-family: 'Segoe UI', Tahoma, sans-serif; font-size: 20px; color: #ffffff; font-weight: bold; padding: 10px; border-radius: 8px; background: linear-gradient(135deg, #2EA149, #60a5fa); text-align: right;"> خواندن دیتای از پیش ایجاد شده </div>


```python
df_income = pd.read_csv('income.csv')
df_income
```

### <div dir="rtl" style="font-family: 'Segoe UI', Tahoma, sans-serif; font-size: 20px; color: #ffffff; font-weight: bold; padding: 10px; border-radius: 8px; background: linear-gradient(135deg, #2EA149, #60a5fa); text-align: right;"> ترکیب داده ها و جمع مقادیر </div>


```python
df_income = df_income.groupby(['Address', 'year']).agg(
    income_w_y_sum=pd.NamedAgg(column='income_w_y', aggfunc='sum'),
    sale_sum=pd.NamedAgg(column='sale', aggfunc='sum'),
    total_other_sum=pd.NamedAgg(column='total_other', aggfunc='sum')
).reset_index()

df_income['income'] = df_income[['income_w_y_sum', 'sale_sum', 'total_other_sum']].sum(axis=1)
```

### <div dir="rtl" style="font-family: 'Segoe UI', Tahoma, sans-serif; font-size: 20px; color: #ffffff; font-weight: bold; padding: 10px; border-radius: 8px; background: linear-gradient(135deg, #2EA149, #60a5fa); text-align: right;"> حذف ستون های اضافی </div>


```python
df_income=df_income.drop(['income_w_y_sum', 'sale_sum', 'total_other_sum'],axis=1)
df_income
```

### <div dir="rtl" style="font-family: 'Segoe UI', Tahoma, sans-serif; font-size: 20px; color: #ffffff; font-weight: bold; padding: 10px; border-radius: 8px; background: linear-gradient(135deg, #2EA149, #60a5fa); text-align: right;"> تصحیح فرمت مقادیر ستون سال </div>


```python
df_income['year'] = df_income['year'].replace({99: 1399, 98: 1398})
```

### <div dir="rtl" style="font-family: 'Segoe UI', Tahoma, sans-serif; font-size: 20px; color: #ffffff; font-weight: bold; padding: 10px; border-radius: 8px; background: linear-gradient(135deg, #2EA149, #60a5fa); text-align: right;"> ترکیب دیتا فریم های مربوط </div>


```python
df_clustering = pd.merge(df_income, df_costs, on=['Address', 'year'],how='outer')
df_clustering
```

### <div dir="rtl" style="font-family: 'Segoe UI', Tahoma, sans-serif; font-size: 20px; color: #ffffff; font-weight: bold; padding: 10px; border-radius: 8px; background: linear-gradient(135deg, #2EA149, #60a5fa); text-align: right;"> مرتب سازی دیتافریم </div>


```python
df_clustering=df_clustering.sort_values(by='year')
df_clustering = df_clustering.reset_index(drop=True)
df_clustering
```

### <div dir="rtl" style="font-family: 'Segoe UI', Tahoma, sans-serif; font-size: 20px; color: #ffffff; font-weight: bold; padding: 10px; border-radius: 8px; background: linear-gradient(135deg, #2EA149, #60a5fa); text-align: right;"> دریافت خروجی دیتافریم ایجاد شده </div>


```python
df_clustering.to_csv('df_clustering',index=False)
```
