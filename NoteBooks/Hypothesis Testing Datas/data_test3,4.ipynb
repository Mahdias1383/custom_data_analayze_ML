```python
import pandas as pd
```

```python
df_income = pd.read_csv('income.csv')
```

```python
df =pd.concat([df_Rp198,df_Rp199,df_Rp11400,df_Rp11401,df_Up198,df_Up199,df_Up11400,df_Up11401])
```

```python
df.head()
```

```python
import numpy as np
```

```python
df.loc[(df['literacy'] == 'illiterate') & (df['degree'].isna()), 'degree'] = 'None'
```

```python
df=df[['Address','degree']]
```

```python
df_income= df_income.drop(['year','member'], axis=1)
```

```python
df_income
```

```python
df_all=pd.merge(df_income,df,on='Address',how='inner')
df_all
```

```python
df_test3 = df_all.groupby(['Address']).agg(
    income_w_y_sum=pd.NamedAgg(column='income_w_y', aggfunc='sum'),
    sale_sum=pd.NamedAgg(column='sale', aggfunc='sum'),
    total_other_sum=pd.NamedAgg(column='total_other', aggfunc='sum'),
    degree=pd.NamedAgg(column='degree', aggfunc='first'),
).reset_index()
df_test3
```

```python
df_test3['income'] = df_test3[['income_w_y_sum', 'sale_sum', 'total_other_sum']].sum(axis=1)
df_test3
```

```python
df_test3=df_test3[['Address','degree','income']]
```
